```python
import os, re
import pandas as pd
from pandas import DataFrame
from numpy import array, sort, argsort, zeros, floor
from openpyxl import load_workbook
import glob
```

```python
path='../result_rf_DD_SF/bal/*.csv'
```

```python
dfl=[]    
files = glob.glob(path)
for a in files:
    try:
        dfl.append(pd.read_csv(a,sep=';'))
    except pd.errors.EmptyDataError:
        print('arquivo vazio!',a)
```

```python
dfl[0]
```

```python
for a in range(len(dfl)):
    dfl[a]=dfl[a].append({'instance':'dfl'+str(a),'capacity':dfl[a].iloc[0]['capacity'],'balanced':dfl[a].iloc[0]['balanced'],'fixsizerf':dfl[a].iloc[0]['fixsizerf'],'freeintervalrf':dfl[a].iloc[0]['freeintervalrf'],'maxtimerf':dfl[a].iloc[0]['maxtimerf'],'bestsol':dfl[a]['bestsol'].mean(),'elapsedtime':dfl[a]['elapsedtime'].mean()}, ignore_index= True)
```

```python
dfl[0]
```

```python
resumedf = pd.DataFrame(columns = ['instance','NR','NT','NW','capacity','balanced','fixsizerf','freeintervalrf','maxtimerf','bestsol','elapsedtime'])
```

```python
for a in range(len(dfl)):
    resumedf=resumedf.append(dfl[a].iloc[3], ignore_index= True)
```

```python
resumedf
```

```python
writer = pd.ExcelWriter('result_bal.xlsx', engine='xlsxwriter')
```

```python
for a in range(len(dfl)):
    dfl[a].to_excel(writer, sheet_name = 'dfl'+str(a))

resumedf.to_excel(writer, sheet_name = 'resume')

writer.save()
```
