```python
esls.columns = [
    'instance',
    'form',
    'cap',
    'capw',
    'capr',
    'T',
    'W',
    'R',
    'bestbound',
    'bestsol',
    'gap',
    'time',
    'nodes',
    'opt'
]
```

```python
rffo.columns = [
    'instance',
    'form',
    'R',
    'T',
    'W',
    'cap',
    'capw',
    'capr',
    'bal',
    'fixsizerf',
    'freeintervalrf',
    'maxtimerf',
    'bestsolrf',
    'timerf',
    'fixsizefo',
    'freeintervalfo',
    'minroundsfo',
    'maxtimefo',
    'incfixfo',
    'inchorfo',
    'bestsolfo',
    'timefo',
    'rounds',
    ]   
```

```python
esls['opt'] = [ 1 if s <= epsilon else 0 for s in esls['gap']]
```

```python
geral = pd.merge(esls,rffo,how='left',on=['instance','cap','capw','capr','T','W','R']) 
```

```python
def gap_rffo(row):
    return (row['bestsolfo'] - row['bestbound'])/(row['bestsolfo'])*100

geral['gap_rffo'] = geral.apply(gap_rffo, axis=1)
```

```python
def improve_fo_rf(row):
    return (row['bestsolrf'] - row['bestsolfo'])/(row['bestsolrf'])*100

geral['impr_fo_rf'] = geral.apply(improve_fo_rf, axis=1)
```

```python
def improve_rffo_esls(row):
    return (row['bestsol'] - row['bestsolfo'])/(row['bestsol'])*100

geral['impr_rffo_esls'] = geral.apply(improve_rffo_esls, axis=1)
```

```python
geral['dif'] = geral['bestsol'] - geral['bestsolfo']
```

```python
geral['#inst<='] = [ 1 if s >= -epsilon else 0 for s in geral['dif']]

geral['#inst<='].sum()
```

```python
geral['#inst<'] = [ 1 if s >= epsilon else 0 for s in geral['dif']]

geral['#inst<'].sum()
```

```python
geral = geral.append({
    'bestsol':geral['bestsol'].mean(),
    'gap':geral['gap'].mean(),
    'opt':geral['opt'].sum(),
    'bestsolrf':geral['bestsolrf'].mean(),
    'bestsolfo':geral['bestsolfo'].mean(),
    'rounds':geral['rounds'].mean(),
    'gap_rffo':geral['gap_rffo'].mean(),
    'impr_fo_rf':geral['impr_fo_rf'].mean(),
    'impr_rffo_esls':geral['impr_rffo_esls'].mean(),
    '#inst<=':geral['#inst<='].sum(),
    '#inst<':geral['#inst<'].sum(),
}, ignore_index= True)
```
